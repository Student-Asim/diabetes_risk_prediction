```python
import pandas as pd
import joblib

# Load model
model_path = "catboost_model.pkl"
loaded_model = joblib.load(model_path)

# Prepare new samples as DataFrame (only features, no target)
new_samples = pd.DataFrame([
    [1, 0, 28.5, 0, 0, 0, 1, 1, 0, 0, 1, 3, 1, 0, 0, 35, 0, 4, 2, 1],
    [0, 1, 32.0, 1, 0, 0, 0, 0, 0, 0, 4, 0, 1, 0, 1, 50, 0, 3, 3, 0]
], columns=['HighBP', 'HighChol', 'BMI', 'Smoker',
            'Stroke', 'HeartDiseaseorAttack', 'PhysActivity', 'Fruits', 'Veggies',
            'HvyAlcoholConsump', 'AnyHealthcare', 'GenHlth', 'MentHlth', 'PhysHlth',
            'DiffWalk', 'Sex', 'Age', 'CardioIssue', 'HealthyLifestyleScore', 'ComorbidityCount'])

# Predict
new_pred = loaded_model.predict(new_samples)
new_pred_proba = loaded_model.predict_proba(new_samples)

# Add predictions to DataFrame
new_samples['Predicted_Class'] = new_pred
new_samples['Prob_Class_0'] = new_pred_proba[:, 0]
new_samples['Prob_Class_1'] = new_pred_proba[:, 1]
new_samples['Prob_Class_2'] = new_pred_proba[:, 2]

new_samples
```
