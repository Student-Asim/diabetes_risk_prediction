4. Feature Engineering:

```python
import pandas as pd
import numpy as np

# Assuming df is your DataFrame

# 1. Combine related binary features: HeartDiseaseorAttack + Stroke → CardioIssue
# If either HeartDiseaseorAttack or Stroke is 1, CardioIssue = 1, else 0
df['CardioIssue'] = np.where((df['HeartDiseaseorAttack']==1) | (df['Stroke']==1), 1, 0)

# 2. BMI categories (raw BMI)
def bmi_std_category(z):
    if z < -1:
        return 'Low BMI'
    elif z > 1:
        return 'High BMI'
    else:
        return 'Average BMI'

df['BMI_category'] = df['BMI'].apply(bmi_std_category)
df['BMI_category'].value_counts()

# 3. Age groups (mapping 1-13 to meaningful ranges)
# Example mapping: 1-4=Young, 5-9=Adult, 10-13=Senior
def age_group(age):
    if age <= 4:
        return 'Young'
    elif age <= 9:
        return 'Adult'
    else:
        return 'Senior'

df['Age_group'] = df['Age'].apply(age_group)

# 4. Lifestyle score (sum of healthy habits)
# Assuming PhysActivity, Fruits, Veggies are 0/1, HvyAlcoholConsump: 1=heavy, 0=not heavy
# Reverse HvyAlcoholConsump so that healthy=1, unhealthy=0
df['HealthyLifestyleScore'] = df['PhysActivity'] + df['Fruits'] + df['Veggies'] + (1 - df['HvyAlcoholConsump'])

# 5. Comorbidity count (number of risk factors per person)
# Sum of HighBP, HighChol, Stroke, HeartDiseaseorAttack
df['ComorbidityCount'] = df[['HighBP','HighChol','Stroke','HeartDiseaseorAttack']].sum(axis=1)

# 6. Interaction features
# Example interactions: BMI * PhysActivity, Age * BMI
df['BMI_x_PhysActivity'] = df['BMI'] * df['PhysActivity']
df['Age_x_BMI'] = df['Age'] * df['BMI']

# Optional: check the new features
df[['CardioIssue','BMI_category','Age_group','HealthyLifestyleScore','ComorbidityCount','BMI_x_PhysActivity','Age_x_BMI']].head()
```

# train the model

1. Using Random Forest Classifier:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

# Initialize
rf = RandomForestClassifier(n_estimators=100, random_state=42)

# Train
rf.fit(X_train, y_train)

# Predict
y_pred_rf = rf.predict(X_test)

# Evaluate
print("Random Forest Classification Report:")
print(classification_report(y_test, y_pred_rf))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_rf))
```

2. XGBOOST

```python
from xgboost import XGBClassifier

# Initialize
xgb = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    use_label_encoder=False,
    eval_metric='mlogloss'
)

# Train
xgb.fit(X_train, y_train)

# Predict
y_pred_xgb = xgb.predict(X_test)

# Evaluate
print("XGBoost Classification Report:")
print(classification_report(y_test, y_pred_xgb))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_xgb))
```

# hyper parameter pruning :

```python
from sklearn.model_selection import GridSearchCV
from xgboost import XGBClassifier

param_grid_xgb = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2],
    'subsample': [0.7, 0.8, 1.0],
    'colsample_bytree': [0.7, 0.8, 1.0]
}

xgb_grid = GridSearchCV(
    estimator=XGBClassifier(random_state=42, eval_metric='mlogloss', use_label_encoder=False),
    param_grid=param_grid_xgb,
    cv=3,
    n_jobs=-1,
    verbose=2,
    scoring='f1_macro'
)

xgb_grid.fit(X_train, y_train)

print("Best parameters:", xgb_grid.best_params_)
best_xgb = xgb_grid.best_estimator_

# Evaluate
y_pred_best_xgb = best_xgb.predict(X_test)
print(classification_report(y_test, y_pred_best_xgb))
```

3. LightGBM

```python
import lightgbm as lgb
from sklearn.metrics import classification_report, confusion_matrix

# Optional: assign class weights to balance classes
from sklearn.utils.class_weight import compute_class_weight
import numpy as np

classes = np.unique(y_train)
class_weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
weight_dict = {cls: w for cls, w in zip(classes, class_weights)}

# Initialize LightGBM
lgbm = lgb.LGBMClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=7,
    class_weight=weight_dict,
    random_state=42
)

# Train
lgbm.fit(X_train, y_train)

# Predict
y_pred_lgbm = lgbm.predict(X_test)

# Evaluate
print("LightGBM Classification Report:")
print(classification_report(y_test, y_pred_lgbm))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_lgbm))
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import classification_report, confusion_matrix

cat = CatBoostClassifier(
    loss_function='MultiClass',
    eval_metric='TotalF1',
    verbose=0,
    random_seed=42,
    class_weights=[1,1,1]
)


param_dist = {
    'depth': [4, 5, 6, 7, 8, 9, 10],
    'learning_rate': [0.01, 0.03, 0.05, 0.1],
    'iterations': [300, 500, 700, 900],
    'l2_leaf_reg': [1, 3, 5, 7, 9],
    'border_count': [32, 64, 128, 254],
    'bagging_temperature': [0, 0.2, 0.5, 1, 2]
}

search = RandomizedSearchCV(
    estimator=cat,
    param_distributions=param_dist,
    n_iter=40,
    scoring='f1_macro',
    cv=3,
    n_jobs=-1,
    verbose=2,
    random_state=42
)

search.fit(X_train, y_train)

print("Best params:", search.best_params_)

best_cat = search.best_estimator_

y_pred = best_cat.predict(X_test)

print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

```python
cat_final = CatBoostClassifier(
    loss_function='MultiClass',
    eval_metric='TotalF1',
    depth=10,
    learning_rate=0.1,
    iterations=500,
    l2_leaf_reg=1,
    border_count=128,
    bagging_temperature=2,
    class_weights=[1,1,1],
    verbose=0,
    random_seed=42
)

cat_final.fit(X_train, y_train)
```

```python
from sklearn.model_selection import cross_val_score
import numpy as np

# 5-fold cross-validation with f1_macro for multi-class
scores = cross_val_score(cat_final, X_train, y_train, cv=5, scoring='f1_macro', n_jobs=-1)

print("F1 Macro scores for each fold:", scores)
print("Mean F1 Macro score:", np.mean(scores))
print("Standard deviation:", np.std(scores))
```
